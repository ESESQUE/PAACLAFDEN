## Creating Normal Data based on Domain Knowledge (Experience) -TYPE 1

```python
#[temp, humidity,aqi, noise, occupancy, co, co2, flame, data_sensitivity]

min_max = ([18, 30], [35, 65], [60, 80], [55, 70], [0, 2],[0,15],[600,1000],[0,1],[0,1])


def make_series(low, high, name):
    if any(isinstance(_, float) for _ in (low, high)):
        func = np.random.uniform
    else:
        func = np.random.randint
    return pd.Series(func(low, high, size=(1000,)), name=name)

column_names = ("temp", "humidity", "aqi", "noise", "occupancy", "co", "co2", "flame", "data_sensitivity")

df1 = pd.concat([make_series(lo, hi, name) for (lo, hi), name in zip(min_max, column_names)],
          axis=1)
```

### Scaling the nodes

```python
def split_data(df, n):
    # Get all unique classes
    all_classes = list(df['data_sensitivity'].unique())

    # Calculate the maximum number of rows per node
    max_rows_per_node = len(df) // n

    # # Initialize a list to hold DataFrame subsets for each node
    # node_subsets = []

    # Initialize a list for train_nodes and test_nodes
    train_nodes = []
    test_nodes = []

    # Create n subsets
    for _ in range(n):
        # Randomly sample subset of all_classes
        subset_classes = random.sample(all_classes, k=random.randint(1, len(all_classes)))

        # Filter DataFrame to only include rows with classes in subset_classes
        node_subset = df[df['data_sensitivity'].isin(subset_classes)]

        # If the node_subset is larger than max_rows_per_node, then randomly select max_rows_per_node
        if len(node_subset) > max_rows_per_node:
            node_subset = node_subset.sample(n=max_rows_per_node)

        # Drop these rows from the original DataFrame
        df = df.drop(node_subset.index)

        # # Add node_subset to list
        # node_subsets.append(node_subset)

        # split node_subset into 80% train and 20% test by taking the first 80% rows for train and the last 20% rows for test
        train_node = node_subset.iloc[:int(len(node_subset)*0.8)]
        test_node = node_subset.iloc[int(len(node_subset)*0.8):]

        train_nodes.append(train_node)
        test_nodes.append(test_node)


    return train_nodes, test_nodes
```
